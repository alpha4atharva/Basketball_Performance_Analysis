```python
import pandas as pd
import numpy as np
```

```python
data = {
    "x1":[4,9,8,8,7,4],
    "x2":[9,8,7,6,5,4],
    "x3":[1,2,3,4,5,6],
    "y":[0,0,0,1,1,1]
}
df = pd.DataFrame(data)
df.head()
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def split(df: pd.DataFrame, train_size):
    train_df = df.sample(frac=train_size/100, random_state=1)
    test_df = df.drop(train_df.index)
    return train_df, test_df
```

```python
def logistic_regression_train(X_train, y_train, lr=0.1, epochs=500):
    X_train = np.c_[np.ones(X_train.shape[0]), X_train]
    beta = np.zeros(X_train.shape[1])

    for epoch in range(epochs):
        z = X_train @ beta
        y_pred = sigmoid(z)
        gradient = X_train.T @ (y_pred - y_train) / len(y_train)
        beta -= lr * gradient

    return beta
```

```python
def logistic_predict(X, beta):
    X = np.c_[np.ones(X.shape[0]), X]
    probs = sigmoid(X @ beta)
    return probs
```

```python
train_df, test_df = split(df, 75)

X_train = train_df[['x1', 'x2', 'x3']].values
y_train = train_df['y'].values
X_test = test_df[['x1', 'x2', 'x3']].values
y_test = test_df['y'].values

beta = logistic_regression_train(X_train, y_train)
probs = logistic_predict(X_test, beta)
preds = (probs >= 0.5).astype(int)

print('Beta coefficients:', np.round(beta, 4))
print('Test probabilities:', np.round(probs, 3))
print('Test predictions:', preds)
print('Actual labels:', y_test)
```
