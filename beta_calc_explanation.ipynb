# How beta is calculated in Logistic Regression (Gradient Descent)

**Algorithm**: Iterative optimization to minimize log-loss.

1. **Initialize**: `beta = np.zeros((n_features + 1))` (intercept included).
2. **For each epoch (1000 iterations)**:
   - `z = X_train @ beta` (linear predictor).
   - `y_pred = sigmoid(z)` (predicted probabilities).
   - **Gradient**: `X_train.T @ (y_pred - y_train) / n_samples` (derivative of log-loss).
   - **Update**: `beta -= lr * gradient` (lr=0.01).
3. **Converges** to beta minimizing loss.

```
def logistic_regression_train(X_train, y_train, lr=0.01, epochs=1000):
    X_train = np.c_[np.ones(X_train.shape[0]), X_train]  # Add intercept
    beta = np.zeros(X_train.shape[1])
    
    for epoch in range(epochs):
        z = X_train @ beta
        y_pred = 1 / (1 + np.exp(-z))  # sigmoid
        gradient = X_train.T @ (y_pred - y_train) / len(y_train)
        beta -= lr * gradient
    
    return beta
```

**Why this gradient?** Log-loss derivative: ∂L/∂β_j = (1/n) Σ (p_i - y_i) * x_ij

**In fixed notebook**: Data generated with true beta ≈ [1, 2, -1]; GD finds close values.
